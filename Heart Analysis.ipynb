```python
heart_data['BMI'].fillna(heart_data['BMI'].mean(), inplace=True)
```

```python
heart_data['PhysicalHealth'].fillna(heart_data['PhysicalHealth'].mean(), inplace=True)
heart_data['MentalHealth'].fillna(heart_data['MentalHealth'].mean(), inplace=True)
heart_data['SleepTime'].fillna(heart_data['SleepTime'].mean(), inplace=True)
```

```python
heart_data.isna().sum()
```

```python
heart_data['Smoking'].fillna('Yes', inplace=True)
heart_data['AlcoholDrinking'].fillna('No', inplace=True)
heart_data['Stroke'].fillna('No', inplace=True)
heart_data['DiffWalking'].fillna('No', inplace=True)
heart_data['Diabetic'].fillna('No', inplace=True)
heart_data['PhysicalActivity'].fillna('No', inplace=True)
heart_data['GenHealth'].fillna('Good', inplace=True)
heart_data['KidneyDisease'].fillna('No', inplace=True)
```

```python
heart_data.isna().sum()
```

```python
heart_data['Sex'].fillna('Male', inplace=True)
heart_data['AgeCategory'].fillna('35-50', inplace=True)
heart_data['Race'].fillna('Medium', inplace=True)
```
